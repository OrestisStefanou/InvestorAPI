## Split to training and test set

```python
import datetime as dt

train_set, test_set = utils.split_data_to_train_and_test(
    df=dataset,
    cutoff_date=dt.datetime(2023,6,1)
)
```

## Split to input and target

```python
X_train, y_train, X_test, y_test = utils.split_into_input_and_target(
    train_set=train_set,
    test_set=test_set
)
```

## Transform input

```python
from sklearn.preprocessing import (
    OneHotEncoder,
    MinMaxScaler
)

one_hot_encoder = OneHotEncoder()
# input_scaler = MinMaxScaler(feature_range=(-1, 1))

X_train_transformed = utils.transform_input(
    X=X_train,
    one_hot_encoder=one_hot_encoder,
    # min_max_scaler=input_scaler,
    fit=True
)

X_test_transformed = utils.transform_input(
    X=X_test,
    one_hot_encoder=one_hot_encoder,
    # min_max_scaler=input_scaler,
    fit=False
)
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

sel_ = SelectFromModel(Lasso())
sel_.fit(X_train_transformed, y_train)
```

## Linear Regression

```python
import matplotlib.pyplot as plt
import pandas as pd

from sklearn import linear_model
from sklearn.metrics import mean_absolute_percentage_error, r2_score, mean_absolute_error

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train_transformed, y_train)

# Make predictions using the testing set
y_pred = pd.Series(regr.predict(X_test_transformed))

# The mean squared error
print("Mean absolute error: %.2f" % mean_absolute_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))
print("Mean absolute pct error: %.2f" % mean_absolute_percentage_error(y_test, y_pred))
```

```python
negative_count = 0
for i, pred_value in enumerate(y_pred):
    if pred_value < 0:
        print(f"Pred value: {pred_value}, actual value: {y_test[i]}")
        negative_count += 1

negative_count
```

## SGDRegressor

```python

# Create linear regression object
sgd_regr = linear_model.SGDRegressor()

# Train the model using the training sets
sgd_regr.fit(X_train_transformed, y_train)

# Make predictions using the testing set
y_pred = pd.Series(sgd_regr.predict(X_test_transformed))

# The mean squared error
print("Mean squared error: %.2f" % mean_absolute_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))

print("Mean absolute pct error: %.2f" % mean_absolute_percentage_error(y_test, y_pred))
```

## Random Forrest

```python
from sklearn.ensemble import RandomForestRegressor

random_forest_regr = RandomForestRegressor(n_estimators=100)

random_forest_regr.fit(X_train_transformed, y_train)

# Make predictions using the testing set
y_pred_rf = pd.Series(random_forest_regr.predict(X_test_transformed))

# The mean squared error
print("Mean absolute error: %.2f" % mean_absolute_error(y_test, y_pred_rf))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred_rf))
print("Mean absolute pct error: %.2f" % mean_absolute_percentage_error(y_test, y_pred_rf))
```
