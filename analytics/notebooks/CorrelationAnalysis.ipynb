## Check for multicollinearity in the data

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
stocks_df_only_numeric = stocks_df.drop(['symbol', 'fiscal_date_ending', 'reported_currency', 'sector', 'price'], axis=1)
```

```python
stocks_df_only_numeric
```

```python
vif_df = calc_vif(stocks_df_only_numeric)
```

```python
vif_df
```

```python
vif_df.sort_values(by=['VIF'], ascending=False, inplace=True)

higher_than_10_list = []
less_or_equal_than_10_list = []
less_or_equal_than_5_list = []

for index, row in vif_df.iterrows():
    if row['VIF'] > 10:
        higher_than_10_list.append((row['variables'], row['VIF']))
    elif row['VIF'] <= 10 and row['VIF'] > 5:
        less_or_equal_than_10_list.append((row['variables'], row['VIF']))
    elif row['VIF'] <= 5:
        less_or_equal_than_5_list.append((row['variables'], row['VIF']))
    else:
        print("What the hell")
```

```python
higher_than_10_list
```

```python
import numpy as np

# Find the strong correlations (> 0.7)
strong_correlations = np.where(np.abs(corr_matrix) > 0.7)

# Create a DataFrame to display the strong correlations
strong_correlations_df = pd.DataFrame({
    'Feature1': final_stocks_df_only_numeric.columns[strong_correlations[0]],
    'Feature2': final_stocks_df_only_numeric.columns[strong_correlations[1]],
    'Correlation': corr_matrix.values[strong_correlations]
})

# Filter out the correlations where Feature1 is not equal to Feature2
strong_correlations_df = strong_correlations_df[strong_correlations_df['Feature1'] != strong_correlations_df['Feature2']]

# Display the strong correlations
strong_correlations_df
```
